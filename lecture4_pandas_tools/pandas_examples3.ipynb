### Example 1 - Removing duplicate data

Usually in ML applications, we want to remove exact duplicates from the dataset, because they might reinforce incorrect trends into the ML model.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("ranges_duplicates.csv")
```

```python
df
```

**We usually want to remove all duplicate data, so our analysis and ML models don't get unrealistic representation for certain rows in data (optimization)**

```python
# we seem to have 3 duplicate rows in the data
df.duplicated().sum()
```

```python
# removing duplicates is luckily easy
df = df.drop_duplicates()
```

```python
# we seem to have 3 duplicate rows in the data
df.duplicated().sum()
```
