### Web scraping, small examples

```python
import json
import numpy as np 
import pandas as pd 
import seaborn as sns

# pip install beautifulsoup4
# pip install requests
from bs4 import BeautifulSoup
import requests
```

```python
# define the URL of the webpage and download the raw HTML
url = "https://en.wikipedia.org/wiki/Rovaniemi"

# modern websites often require headers set correctly before 
# we can remotely read the website contents (HTML)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}

# load the page data with headers
page = requests.get(url, headers=headers)
```

```python
# create a "soup object" of the new raw web page
soup = BeautifulSoup(page.content, "html.parser")
```

```python
# let's find something within the data

# NOTE! use developer tools in your web browser (usually F12)
# and find the details regarding what you wish to scrape
# in this case => <h1> -tag with the id of firstHeading + get 
# the text content of this element
cityname = soup.find("h1", id="firstHeading").text
cityname

# if targeting class => class_
# if targeting is => id
```

```python
# getting multiple elements, <tr> with class "mergedtoprow"
# this is the small summary table on the right side of the page
elements = soup.find_all("tr", class_="mergedtoprow")
```

```python
# loop through all that matched with <tr> and class="mergedtoprow"
for e in elements:
    # you can use if-statements to filter the particular value
    # you want from this (see later example)
    print(e.text)
```

```python
# get the municipality, a <span> with the title "English-language text"
municipality = soup.find("span", title="English-language text").text
municipality
```

```python
caption_text = soup.find("div", class_="ib-settlement-caption").text
caption_text
```

```python
elements = soup.find_all("tr", class_="mergedtoprow")

# 15 elements, 31.3.2026
# print(len(elements))

# loop through all 15 elements
for e in elements:
    # print(e.text)

    # all mergedtoprows don't have <th> => avoid with if-statement
    if e.find("th") != None:
        # if there IS a <th> => check if the title is "Time zone"
        if e.find("th").find(string=True) == "Time zone":
            # since this seems to be the correct slot
            # => select next element that is <td> => this contains the text we want
            info = e.find_next("td")
            print(info.text)

            # we found the value we wanted => break and stop loop
            break
```
