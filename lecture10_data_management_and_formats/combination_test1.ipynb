### Combining DataFrames, example 1 - simple concatenation

```python
import pandas as pd

# our test data (two DataFrames) for concatenation
products1 = [
    {"product_id": 11, "name": "SomeCorp CPU Accessory Kit 3", "price": 150},
    {"product_id": 12, "name": "WatIsDis Cooling Gear Collection 2024", "price": 250},
    {"product_id": 13, "name": "Generyc Processor f5-A12", "price": 350},
    {"product_id": 14, "name": "SemiPremium Processor f7-B33", "price": 449},
    {"product_id": 15, "name": "Ultimate Processor f9-C99", "price": 550},
]

products2 = [
    {"product_id": 16, "name": "ThunderStorm PSU 850W DX850", "price": 175},
    {"product_id": 17, "name": "KindaSUS Motherboard NL-175", "price": 273},
    {"product_id": 18, "name": "DaBasicGPU OfficeBlaster 1500-XT ", "price": 377},
    {"product_id": 19, "name": "Gaming GPU XBlazer 6700-XY", "price": 675},
    {"product_id": 20, "name": "AI-trainerGPU cardTX-2005", "price": 1279},
]

df1 = pd.DataFrame(products1)
df2 = pd.DataFrame(products2)
```

```python
# use concat() to combine these DataFrames into a single large DataFrame
df_combined = pd.concat([df1, df2])

# reset the index se we don't have duplicate ids from the original DataFrames
df_combined = df_combined.reset_index()

# drop the original index (that is wrong)
df_combined = df_combined.drop("index", axis=1)
```
