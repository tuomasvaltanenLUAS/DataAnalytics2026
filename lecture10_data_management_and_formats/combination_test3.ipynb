### Combining DataFrames, example 3 - more complex merge

```python
import pandas as pd

products = [
    {"product_id": 11, "name": "SomeCorp CPU Accessory Kit 3", "price": 150},
    {"product_id": 12, "name": "WatIsDis Cooling Gear Collection 2024", "price": 250},
    {"product_id": 13, "name": "Generyc Processor f5-A12", "price": 350},
    {"product_id": 14, "name": "SemiPremium Processor f7-B33", "price": 449},
    {"product_id": 15, "name": "Ultimate Processor f9-C99", "price": 550}
]

orders = [
    {"product_id": 11, "delivery_time": 5, "delivery_method": "postal"},
    {"product_id": 11, "delivery_time": 3, "delivery_method": "courier"},
    {"product_id": 11, "delivery_time": 6, "delivery_method": "postal"},
    {"product_id": 12, "delivery_time": 7, "delivery_method": "postal"},
    {"product_id": 12, "delivery_time": 4, "delivery_method": "courier"},
    {"product_id": 12, "delivery_time": 8, "delivery_method": "postal"},
    {"product_id": 13, "delivery_time": 4, "delivery_method": "courier"},
    {"product_id": 13, "delivery_time": 6, "delivery_method": "postal"},
    {"product_id": 13, "delivery_time": 7, "delivery_method": "postal"},
    {"product_id": 14, "delivery_time": 8, "delivery_method": "postal"},
    {"product_id": 14, "delivery_time": 5, "delivery_method": "courier"},
    {"product_id": 14, "delivery_time": 6, "delivery_method": "postal"}
]

df1 = pd.DataFrame(products)
df2 = pd.DataFrame(orders)
```

```python
df1
```

```python
df2
```

### If we just merge the DataFrames like this, we'll mess up the data usually

... because there are more than one rows of data for each product in the second DataFrame.

```python
df_new = df1.merge(df2, left_on="product_id", right_on="product_id", how="left")
```

```python
df_new
```

```python
# in most cases, you don't want the amount of rows to increase
# after joining data!
print(f"Amount of data before joining: {len(df1)}")
print(f"Amount of data AFTER joining: {len(df_new)}")
```

### Usually we have to reduce the amount of rows in the second DataFrame, so it matches the first one => use averages and groupby()

```python
# group the second DataFrame by averaging the delivery time
# for each product!
df2_grouped = df2.groupby("product_id").mean(numeric_only=True)
```

```python
# grouped average delivery times
df2_grouped
```

```python
# merge with the grouped version of the second DataFrame
df_new = df1.merge(df2_grouped, left_on="product_id", right_on="product_id", how="left")
```

```python
# data amounts seem fine!
print(f"Amount of data before joining: {len(df1)}")
print(f"Amount of data AFTER joining: {len(df_new)}")
```

```python
# merging looks OK, notice how one delivery time is NaN
# since the original df2 did not have any delivery times for product_id = 15
df_new
```
