```python
import xml.etree.ElementTree as et

# load XML-file -> get root node
xtree = et.parse("somedata_single.xml")
node = xtree.getroot()

# column names for pandas when we create the final DataFrame
df_cols = ["id", "name", "value"]
rows = []

# parse "columns" or values in the XML
e_id = node.find("id").text
e_name = node.find("name").text
e_value = node.find("value").text

# create the DataFrame
rows.append({'id': e_id, 'name': e_name, 'value': e_value})
out_df = pd.DataFrame(rows, columns=df_cols)
```

```python
import pandas as pd
import xml.etree.ElementTree as et 
xtree = et.parse("somedata.xml")

# in this data, we have two nodes
nodes = xtree.getroot()
df_cols = ["id", "name", "value"]
rows = []

# loop through all item in array
for node in nodes:
    e_id = node.find("id").text
    e_name = node.find("name").text
    e_value = node.find("value").text

    rows.append({"id": e_id, "name": e_name, "value": e_value})
 
# create the DataFrame 
out_df = pd.DataFrame(rows, columns = df_cols)
```

```python
out_df
```

```python
# IF YOU CAN: TRY TO ALWAYS USE pd.read_xml() and the xpath-parameter
# in very very few cases, you might need to revert back to the traditional
df = pd.read_xml("somedata_single.xml", xpath="/*")
df
```

```python
# this is the default "xpath" in xml
# basically => data is the root node => use everything "*" inside data as an ITEM
df = pd.read_xml("somedata.xml", xpath="/data/*")
df
```
