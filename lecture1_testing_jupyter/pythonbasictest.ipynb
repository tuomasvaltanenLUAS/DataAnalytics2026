### Conditionals etc.

```python
# we usually don't use input() in Jupyter notebooks
# because it slows the notebook down and we often
# want to run all code cells at once
number = 55

# conditionals are usually used in custom filtering functions
# especially in pandas-module
if number > 0:
    print("BIG NUMBER!")
elif number == 0:
    print("ZERO!")
else:
    print("Something else...")
```

```python
# sometimes we might need to do something like this in text AIs
text = "hellooooooooooo there!"

# text comparisons more in NLP AIs (natural language processing)
if "hello" in text:
    print("HELLO detected!")
else:
    print("Nevermind")
```

### Collections, lists etc.

```python
# imagine you have a photo classification AI
# supported categories => dog, cat, fox
categories = ["dog", "cat", "fox"]
```

```python
# this could be the result from a photo detection AI model
result_index = 1

# and typically we can use the list above the for the name of the category
print(categories[result_index])
```

```python
# indexes in lists
temperatures = [-29, -32, -15, -12, 5]
temperatures[2]
```

```python
# in NumPy and pandas, we'll have new versions of this slicing syntax
temperatures[1:3]
```

```python
# dictionaries are often needed when you have trained
# an AI to detect something
example_data = {
    "area": 132,
    "bedrooms": 5,
    "bathrooms": 2,
    "sauna": 1
}

# then you process the dictionary into the format expected by AI
# => and get a result (a price for example)
example_data
```

```python
# we quite rarely need typical numeric loops in AI or data analytics...
for x in range(2020, 2027):
    print(x)
```
