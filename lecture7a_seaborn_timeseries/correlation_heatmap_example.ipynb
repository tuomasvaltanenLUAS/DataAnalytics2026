### One common usage for heatmaps is to visualize a correlation matrix quickly

```python
import pandas as pd
import seaborn as sns
```

```python
# load the usual tips -dataset
df = sns.load_dataset("tips")
```

```python
# get correlations into a matrix
correlations = df.corr(numeric_only=True)
correlations
```

```python
# visualize correlations, however this data is very simple
# so the visualization doesn't really add that much here
sns.heatmap(correlations, annot=True)

# see also the example with the housing data and see what happens with more columns
# => lecture 5 => house analysis version 1
```
