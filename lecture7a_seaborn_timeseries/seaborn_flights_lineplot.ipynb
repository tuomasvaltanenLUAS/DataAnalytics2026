### While line plots are quite practical with time series data, it's not always the most useful option

**Nevertheless, let's try with flights -data first!**

Remember to check the Delhi -example for a more useful multiline plot!

```python
import seaborn as sns
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
# with a basic line plot and one variable, we get a very generic trend
# => amount of passengers increase along the years
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
# with the hue, we can compare monthly differences from year to year
# however, this is not very easy to read...?
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```

```python
# after swapping the month and year, we can see somewhat the monthly distribution
# on summer time, but it's still not as clear as it could be
sns.lineplot(data=flights, x="month", y="passengers", hue="year")
```

**Rule of thumb: multiline plots are best with time series data where the variables are DIFFERENT variables.**

... with this flights data, all the variables are related to each other, and it's hard to see trends.
