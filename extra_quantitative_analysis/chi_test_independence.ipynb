<b>chi-test: see if the columns are dependent </b>

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.DataFrame({'Season' : ['Winter', 'Winter', 'Winter', 'Summer', 'Summer'] * 10,
                   'AvgTemp' : ['Cold', 'Cold', 'Warm', 'Warm', 'Warm'] * 10
                  })
df
```

```python
# calculate a contingency table to see how the sums
# distribute between the combinations of the values
contigency = pd.crosstab(df['Season'], df['AvgTemp'])
contigency
```

```python
# convert the contingencyies into percentage
contigency_pct = pd.crosstab(df['Season'], df['AvgTemp'], normalize='index')
contigency_pct
```

```python
# Chi-square test of independence.
# value close to 0 implies the two groups are dependent on each other
c, p, dof, expected = chi2_contingency(contigency)
f"{p:.15f}"
```
