```python
# pip install pingouin
import numpy as np
import pandas as pd

# create data
df = pd.DataFrame({'technique': np.repeat(['Self-study', 'Video lessons', 'Classroom'], 5),
                   'current_grade': [67, 88, 75, 77, 85,
                                     92, 69, 77, 74, 78, 
                                     96, 91, 88, 82, 80],
                   'exam_score': [77, 79, 72, 74, 69,
                                  78, 88, 83, 84, 90,
                                  85, 89, 83, 88, 95]})

# view data 
df
```

```python
# pip install pingouin
from pingouin import ancova

# perform ANCOVA
ancova(data=df, dv='exam_score', covar='current_grade', between='technique')

# p-unc = p-value uncorrected, our p-value
# in most cases, p-values under 0.05 indicate statistical significance
# in this case, if np2 goes up, p-unc probably down

# np2 => effect% of the column towards the target (current grade)
# the other's are measuring the same thing or just variables used by the ancova equation
```
