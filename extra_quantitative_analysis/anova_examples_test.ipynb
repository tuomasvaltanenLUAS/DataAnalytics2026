```python
# pip install scipy

# enter exam scores for each group
# three different groups with three different study methods
# our question = does the study method affect the exam score?
group1 = [60, 76, 58, 65, 78, 64, 68, 59, 61, 70]
group2 = [91, 92, 93, 85, 87, 84, 82, 88, 95, 96]
group3 = [75, 72, 77, 81, 68, 75, 73, 72, 79, 81]
```

```python
from scipy.stats import f_oneway

# perform one-way ANOVA
result = f_oneway(group1, group2, group3)

# this implies that the different groups had different results
# for example, three groups with different study methods
# seem to yield different exam scores
# p-value is = 0.000000000199
print(f"{result.pvalue:.12f}")
```
