<b>We can also use chi-test to examine the goodness-of-fit</b>

```python
# importing packages 
import scipy.stats as stats 
import numpy as np 
  
# number of hours a student studies 
# in a week vs expected number of hours

# basic chi-test has a limitations:
# each value in each list has to have a pair somewhere in the other list
# for example, original training data => trained ML model => how well the model represents the original data
# goodness-of-fit values over 0.9 are good  
observed_data = [8, 8, 10, 9, 8, 11, 9] 
expected_data = [9, 8, 11, 8, 10, 9, 8] 
  
# Chi-Square Goodness of Fit Test 
chi_square_test_statistic, p_value = stats.chisquare( 
    observed_data, expected_data) 
  
# chi square test statistic and p value 
print('chi_square_test_statistic is : ' +
      str(chi_square_test_statistic)) 
print('p_value : ' + str(p_value)) 
  
# find Chi-Square critical value 
print(stats.chi2.ppf(1-0.05, df=6))

# since the plain chi-square goodness of fit is quite limited, we luckily have other tools in ML
# to make this easier, SelectKBest, R-squared etc.
```
