### seaborn - pair plots

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# seaborn datasets also available, originates from here:
# https://github.com/mwaskom/seaborn-data
# these are just csv-files in the background => pandas DataFrames
tips = sns.load_dataset("tips")

# since tips is just a DataFrame, we can add more variables
tips['total_cost'] = tips['total_bill'] + tips['tip']
tips
```

### Pair plot compares every variable with every variable (only numerical)

**Pair plots are usually useful for finding the most common and obvious connections between certain variables QUICKLY. The idea is to use further plots then and inspect these obvious variables in more detail (box plots, regression plots etc.)**

```python
sns.pairplot(tips)
```

### Notice how we have two comparison for each pair (axis flipped), we can remove the other half since the information is redundant (easier to read)

```python
# use corner=True to remove the upperhalf so we don't have redundant information
sns.pairplot(tips, corner=True)
```

```python
# use corner=True to remove the upperhalf so we don't have redundant information
# in this case, we can roughly see that bigger groups result into bigger tips and total bills
# but not much in addition to that
sns.pairplot(tips, corner=True, hue="size")
```

```python
# this is fairly "all over the place" / chaotic, can't really see much in the scatter plots here
sns.pairplot(tips, corner=True, hue="day")
```
