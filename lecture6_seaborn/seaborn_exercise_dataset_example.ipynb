### Box plot -example with the exercise-dataset (from seaborn)

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# load the dataset
df = sns.load_dataset("exercise")
df.heaD()
```

#### This is a very clean box plot example where we can see the differences nicely

```python
# box plot measuring pulse against time, separated by "kind"
# => rest, walking, running
sns.boxplot(x="time", y="pulse", data=df, hue="kind")
```
