### Popular educational dataset: iris (flowers)

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# load the dataset
df = sns.load_dataset("iris")
```

### Scatter plot, we can see some rough linear correlation going (positive)

```python
# scatter plot without the hue
sns.scatterplot(x="sepal_length", y="petal_length", data=df)
```

**With HUE, we can see some clustering between the flower species, the actual data wasn't as simple as we thought with the previous scatter plot!**

```python
# scatter plot WITH the hue
sns.scatterplot(x="sepal_length", y="petal_length", data=df, hue="species")
```

**... but with the hue, we can actually see the different species clustering separately quite clearly**

```python
# we can see some nice separation between the species with MINIMAL overlap
sns.pairplot(df, corner=True, hue="species")
```

### Correlations and basic regression plots can be deceiving...

```python
# this implies: when sepal width increases, then sepal length decreases
# ... WHICH MAKES NO SENSE?
sns.lmplot(x="sepal_width", y="sepal_length", data=df)
```

### This is why regression plot + hue is powerful, we can actually see the underlying connection here.

```python
# so the regression is actually the other way around
# as length increases, width increases (as it makes sense)
# it just is that we have three different species and the
# original regression was confused because of that
sns.lmplot(x="sepal_width", y="sepal_length", data=df, hue='species')
```

```python
# in statistics, this phenomenon above is known as the "Simpson's Paradox"
# https://www.statsig.com/perspectives/simpsons-paradox-explained

# Wiki definition:
# Simpson's paradox is a phenomenon in probability and statistics in which a 
# trend appears in several groups of data but disappears or reverses when 
# the groups are combined."
```
