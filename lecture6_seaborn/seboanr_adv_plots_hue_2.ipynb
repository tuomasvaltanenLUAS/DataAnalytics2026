#### seaborn, advanced plots and features - the hue-feature

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# seaborn datasets also available, originates from here:
# https://github.com/mwaskom/seaborn-data
# these are just csv-files in the background => pandas DataFrames
tips = sns.load_dataset("tips")

# since tips is just a DataFrame, we can add more variables
tips['total_cost'] = tips['total_bill'] + tips['tip']
tips
```

### One of the most powerful features in seaborn -> hue -parameter

**Hue allows us to compare distributions between nominal categories at the same time.**

### Hue is ESPECIALLY powerful in later advanced plots like pair plots, scatter plots, regression plots and box plots!

```python
# for example with barplot, place a categorical variable in x and hue
# and a continuous variable (like price, cost, tip etc.) into y
# this results into a visualization, where we can inspect the distributions/averages
# upon multiple categories at once
sns.barplot(x="sex", y="total_cost", data=tips, hue="day")
```

```python
# available colors are based on matplotlib:
# https://matplotlib.org/stable/users/explain/colors/colormaps.html

# if you want to alter the default color scheme in seaborn, use palette
# in some plots => cmap
sns.barplot(x="sex", y="total_cost", data=tips, hue="day", palette="magma")
```
