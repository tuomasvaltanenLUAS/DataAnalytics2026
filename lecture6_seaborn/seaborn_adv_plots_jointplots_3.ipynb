### seaborn, more advanced plots and features - joint plot

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# seaborn datasets also available, originates from here:
# https://github.com/mwaskom/seaborn-data
# these are just csv-files in the background => pandas DataFrames
tips = sns.load_dataset("tips")

# since tips is just a DataFrame, we can add more variables
tips['total_cost'] = tips['total_bill'] + tips['tip']
tips
```

### Joint Plot => allows combining scatter plot + histograms for both variables

```python
# joint plot => scatter plot + histogram
sns.jointplot(x="total_bill", y="tip", data=tips)
```

### Extra feature: you can also change the scatter plot to a HEX plot

```python
# you can nicely see the most average area of values 
# between these variables with the hex plot
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
```

```python
# also a regression version is possible, shows the general trend line based on the points
sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg")
```

**You can use hue with join plot, BUT ONLY WITH THE BASIC VERSION (unfortunately)**

```python
# joint plot with hue, only works with basic joint plot
sns.jointplot(x="total_bill", y="tip", data=tips, hue="time")
```
