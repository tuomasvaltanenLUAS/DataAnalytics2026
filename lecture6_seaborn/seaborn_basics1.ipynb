### Basic plot types, seaborn

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# seaborn datasets also available, originates from here:
# https://github.com/mwaskom/seaborn-data
# these are just csv-files in the background => pandas DataFrames
tips = sns.load_dataset("tips")

# since tips is just a DataFrame, we can add more variables
tips['total_cost'] = tips['total_bill'] + tips['tip']
tips
```

**Histogram -> distribution of values within a column (in your dataset)**

We often aim for a rough standard normal distribution (real data is rarely an exact normal distribution)

```python
# basic histogram => show the distrbution
# often used to check if the data is in optimal distribution or shape
# often a rough normal distribution or sometimes uniform distribution is optimal
sns.histplot(tips['total_bill'])
```

```python
# let's also try the new variable
sns.histplot(tips['total_cost'])
```

```python
# let's also try the new variable
sns.histplot(tips['total_cost'], kde=True)
```

```python
sns.histplot(tips['total_cost'], kde=True, bins=40)
```
