### Basic plot types, seaborn

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# seaborn datasets also available, originates from here:
# https://github.com/mwaskom/seaborn-data
# these are just csv-files in the background => pandas DataFrames
tips = sns.load_dataset("tips")

# since tips is just a DataFrame, we can add more variables
tips['total_cost'] = tips['total_bill'] + tips['tip']
tips
```

**Histogram -> distribution of values within a column (in your dataset)**

We often aim for a rough standard normal distribution (real data is rarely an exact normal distribution)

```python
# basic histogram => show the distrbution
# often used to check if the data is in optimal distribution or shape
# often a rough normal distribution or sometimes uniform distribution is optimal
sns.histplot(tips['total_bill'])
```

```python
# let's also try the new variable
sns.histplot(tips['total_cost'])
```

```python
# KDE line is handy for seeing the overall trend more smoothly
# NOTE! Can show an incorrect trend shape if not enough data
# available or variance is too hight (data is chaotic)
sns.histplot(tips['total_cost'], kde=True)
```

```python
# you can also adjust the bins: amount of bars shown in the histogram
# the more bins, the more detail shown
# the default bins-amount often shows a very rough simplified distribution
sns.histplot(tips['total_cost'], kde=True, bins=40)
```

```python
# EXTRA: you can also remove the border from each bar/bin, end result is a bit smoother
sns.histplot(tips['total_cost'], kde=True, bins=40, edgecolor="k", linewidth=0)
```

### scatter plots - compare two variables to each other and their relative distribution

```python
# scatter plot, compare two variables into each other
# in this case: total_bill => tip
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
# in the plot above we can see a "rough line from bottom left to top right" 
# => this implies POSITIVE CORRELATION (seems to correlate approximately ~0.68)
# a rough line from TOP LEFT to bottom right => negative correlation
# if points are all over the place => no correlation or connection

# you will sometimes also see data grouping in a seasonality
# => non-linear connection => correlations are usually 0
tips.corr(numeric_only=True)
```
