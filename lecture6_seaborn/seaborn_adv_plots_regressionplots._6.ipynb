### Regression plots are one of the most powerful plots in seaborn

**With regression plots, we can efficiently study our datasets for trends, contradictions, overlaps etc. problems in the dataset.**

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# seaborn datasets also available, originates from here:
# https://github.com/mwaskom/seaborn-data
# these are just csv-files in the background => pandas DataFrames
tips = sns.load_dataset("tips")

# since tips is just a DataFrame, we can add more variables
tips['total_cost'] = tips['total_bill'] + tips['tip']
tips.head()
```

### Regression plots are optimal for finding hidden connections between numeric values and categories


**Usually when correlations fail to show a connection in data, regression plots can show a little bit more about the "truth" behind zero correlations (sometimes data doesn't just have any correlations though, but if there is => regression plot helps)**

```python
# basic regression plot, no hue
sns.lmplot(x="total_bill", y="tip", data=tips)
```

**By default, we usually want to use regression plots with hue, if possible!**

```python
# hue also supported!
# NOTE: you can use with lmplot, but not with the original regplot!
sns.lmplot(x="total_bill", y="tip", data=tips, hue="time")

# not much difference between lunch/dinner though!
```

```python
# let's inspect the smoker status too
# okay, here we can see a little different between the trends
# so it seems non-smokers tip more heavily usually
# however, the confidence level in the smoker-category is pretty bad
# which means the estimation is a bit unsures (probably due to not enough data)
sns.lmplot(x="total_bill", y="tip", data=tips, hue="smoker")
```

```python
# not much difference here going on, same trends
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex")
```

### EXTRA: Create multiple regression plots at once

```python
# compare gender trends but for each day separately
sns.lmplot(x="total_bill", y="tip", data=tips, col="day", hue="sex")
```

```python
# this gets pretty wild...
# in this case, we compare lunch/dinner to each day and also between genders
# here we can also see how lunch is not server on Saturday and Sunday
sns.lmplot(x="total_bill", y="tip", data=tips, col="day", hue="sex", row="time")
```
