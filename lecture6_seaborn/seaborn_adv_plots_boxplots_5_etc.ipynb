### box plots and its variants - plot type to see differences between categorical variables and their distributions

**Box plots are very powerful in analysis - you can easily find trends and contradictions in dataset with these!**

```python
# basic imports
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# seaborn datasets also available, originates from here:
# https://github.com/mwaskom/seaborn-data
# these are just csv-files in the background => pandas DataFrames
tips = sns.load_dataset("tips")

# since tips is just a DataFrame, we can add more variables
tips['total_cost'] = tips['total_bill'] + tips['tip']
tips
```

### Box plots are useful when comparing categories and numeric values and their distributions

**How to read box plot:** https://www.statisticshowto.com/probability-and-statistics/descriptive-statistics/box-plot/

```python
# basically you can see the min and max -values and the majority of the distribution BOX in the middle
# the small circles outside the min and max are OUTLIERS
# the line in the box is the median

# in other words, many statistical features in one picture
```

```python
# basic box plot'
# we can roughly see the total bill tends to go towards Sunday
# Saturday has a lot of variation in data
# meaning that some of the lowest and highest values happen in Saturday
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
# HUE also supported
# here we can again see that lunch is only server on Thursday and Friday
# it seems there's one dinner sold in Thursday as well
sns.boxplot(x="day", y="total_bill", data=tips, hue="time")
```

```python
# male usually spend more money on Friday (probably a popular date night in the restaurant)
# and female slightly more (on average) on Saturday
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
# it seems smokers use more money on the bill during Saturday and Sunday
# but! lots of noise / variation in data (probably due to small dataset size, only 244 rows)
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
```

```python
# let's try one more
# okay, this gets a bit difficult to read

# but we can see here => Friday consists mostly of 2 people groups
# => probably a common date night in this particular restaurant
# also seems Saturday is the most popular day with biggest sales 
# => many groups of 4 visit the restaurant

# Sunday seems to have some extreme outliers (one group of 6 spending almost 50 bucks)
# and this might be the reason we originally thought Sunday was the best selling day
# (and this is why we usually wish to remove extreme outliers from our data)
sns.boxplot(x="day", y="total_bill", data=tips, hue="size")
```
